### Handling Null Vales

```python
results_sample.isna().sum()
```

```python
results_sample[results_sample['place'].isna()]
```

```python
results_sample[results_sample[['place', 'medal']].isna().any(axis=1)]
```

```python
results_sample[results_sample['place'].notna()]
```

```python
results_sample = results.sample(100, random_state = 40)
results_sample.fillna(10)
```

```python
results_sample = results.sample(100, random_state = 40)
results_sample.fillna(results_sample['place'].mean())
```

```python
results_sample = results.sample(100, random_state = 40)
results_sample.fillna(results_sample['place'].interpolate())
```

```python
results_sample = results.sample(100, random_state = 40)
results_sample.dropna()
```

```python
# Drop specific rows
results_sample = results.sample(100, random_state = 40)
results_sample.dropna(subset = ['place'],inplace=True)
results_sample
```

```python
results_sample = results.sample(100, random_state = 40)
results_sample.bfill()
results_sample
```

```python
#Pivot
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python
df
```

```python
pivot_table = df.pivot_table(values=iris.feature_names, 
                             index='species', 
                             aggfunc='mean')
pivot_table
```
