```python
import yfinance as yf
import pandas as pd

# Download historical stock data for Apple (AAPL)
data = yf.download('AAPL', start='2023-01-01', end='2023-12-31')
data
```

```python
df = data[['Close']]
df
```

### Shift

```python
df['Previous_Close'] = df['Close'].shift(1)
df
```

```python
df['Grwoth'] = (((df['Close'] - df['Previous_Close']) / df['Previous_Close'])*100).round(2)
df
```

#### Rank

```python
# Rank the closing prices within the dataset
df['Rank'] = df['Close'].rank()
df
```

#### Rolling

```python
df['7-Day_Rolling_Mean'] = df['Close'].rolling(window=7).mean()
df.head(7)
```

#### Cumsum

```python
# Calculate the cumulative sum of the closing prices
df['Cumulative_Sum'] = df['Close'].cumsum()
df
```
