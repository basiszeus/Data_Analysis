### Filters

```python
sample2 = bios.sample(100)
sample2.info()
```

```python
sample2.loc[sample2['height_cm'] > 190]
```

```python
sample2.loc[(sample2['height_cm'] > 190) & (sample2['weight_kg'] > 100)]
```

```python
sample2.loc[(sample2['height_cm'] == 196) | (sample2['weight_kg'] == 86)]
```

```python
sample2.loc[~(sample2['height_cm'] > 190)]
```

```python
sample2.loc[~(sample2['height_cm'] > 190), ['name', 'height_cm']]
```

```python
sample2.loc[(sample2['height_cm']>160) & (sample2['name'].str.contains('Lefa'))]
```

```python
sample2.loc[sample2['name'].str.contains('Lefa'), 'height_cm']
```

```python
sample2.loc[sample2['name'].str.contains('Lefa'), ['height_cm', 'name']]
```

```python
sample2.loc[sample2['name'].str.contains('Lefa|Erni'),['height_cm', 'name']]
```

```python
sample2.loc[sample2['name'].str.contains('Lefa|Erni', case=False, regex=False), ['height_cm', 'name']]
```

```python
sample2[sample2['name'].isin(['Lefa Tsapi'])]
```

```python
sample2.query("born_country == 'USA' and height_cm > 190")
```
