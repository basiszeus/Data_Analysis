### Categories

```python
bios_sample['weight_category'] = bios_sample['weight_kg'].apply(lambda x:'Light' if x < 50 else ('Average' if x < 75 else 'Heavy'))
bios_sample
```

```python
def category_weight(row):
    if row ['weight_kg'] <50:
        return 'Light'
    elif row ['weight_kg'] <75:
        return 'Average'
    else:
        return 'Heavy'
    
bios_sample['function_demo'] = bios_sample.apply (category_weight, axis =1)
bios_sample
```
